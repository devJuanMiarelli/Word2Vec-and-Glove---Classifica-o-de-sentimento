```python
import gensim.downloader as api

# Carregar embeddings Word2Vec e GloVe pré-treinados
w2v_model = api.load("word2vec-google-news-300")
glove_model = api.load("glove-wiki-gigaword-100")

# Obter vetor de palavra para uma palavra específica
word_vector1 = w2v_model['fuel']
print("Vetor Word2Vec para a palavra 'fuel':", word_vector1)
word_vector2 = glove_model['fuel']
print("Vetor GloVe para a palavra 'fuel':", word_vector2)

# Encontrar palavras mais similares
similar_words1 = w2v_model.most_similar('fuel', topn=5)
print("Palavras mais similares a 'fuel' utilizando a Word2Vec:", similar_words1)
similar_words2 = glove_model.most_similar('fuel', topn=5)
print("Palavras mais similares a 'fuel' utilizando a GloVe:", similar_words2)
```
